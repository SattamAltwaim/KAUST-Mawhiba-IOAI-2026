## The Math Behind Your Model

You will implement each formula below in NumPy.

### 1. Prediction

Given features $\mathbf{X} \in \mathbb{R}^{N \times d}$ and weights $\mathbf{w} \in \mathbb{R}^d$:

$$
\hat{\mathbf{y}} = \mathbf{X} \mathbf{w}
$$

### 2. Loss (Mean Squared Error)

$$
J(\mathbf{w}) = \frac{1}{N} \|\mathbf{X}\mathbf{w} - \mathbf{y}\|^2 = \frac{1}{N} \sum_{i=1}^{N} (\hat{y}_i - y_i)^2
$$

### 3. Gradient

The derivative of $J$ with respect to $\mathbf{w}$:

$$
\nabla_{\mathbf{w}} J = \frac{2}{N} \mathbf{X}^\top (\mathbf{X}\mathbf{w} - \mathbf{y})
$$

### 4. Weight update

Move opposite to the gradient:

$$
\mathbf{w} \leftarrow \mathbf{w} - \alpha \, \nabla_{\mathbf{w}} J
$$
