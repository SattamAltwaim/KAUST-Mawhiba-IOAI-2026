## Part 1: Load Embeddings and Select Words

````python
import gensim.downloader as api

print("Loading GloVe vectors...")
glove = api.load("glove-wiki-gigaword-50")
print(f"Loaded {len(glove)} vectors of dimension {glove.vector_size}")
````

````python
categories = {
    "Animals":   ["cat", "dog", "fish", "bird", "horse", "lion", "tiger",
                  "bear", "wolf", "eagle", "shark", "whale", "rabbit", "deer"],
    "Countries": ["france", "germany", "japan", "china", "brazil", "india",
                  "canada", "australia", "egypt", "mexico", "italy", "spain"],
    "Sports":    ["football", "basketball", "tennis", "swimming", "boxing",
                  "golf", "cricket", "hockey", "volleyball", "baseball"],
    "Colors":    ["red", "blue", "green", "yellow", "black", "white",
                  "purple", "orange", "pink", "brown", "gray"],
    "Verbs":     ["run", "walk", "eat", "sleep", "read", "write", "think",
                  "speak", "build", "drive", "fly", "swim", "dance", "sing"],
}
````

````python
words = []
vectors = []
labels = []
colors_map = {"Animals": "red", "Countries": "blue", "Sports": "green",
              "Colors": "purple", "Verbs": "orange"}

for category, word_list in categories.items():
    for word in word_list:
        if word in glove:
            words.append(word)
            vectors.append(glove[word])
            labels.append(category)

X = np.array(vectors)
print(f"Collected {len(words)} words across {len(categories)} categories")
print(f"Matrix shape: {X.shape}  (words x dimensions)")
````

## Wrap-Up

**Key takeaways:**

- **PCA** finds linear directions of maximum variance -- fast, interpretable, gives explained variance.
- **t-SNE** preserves local neighborhoods -- great for visualization, but slow and non-deterministic.
- **UMAP** is fast, preserves more global structure, and can also be used for preprocessing.
- **Best practice:** PCA to denoise, then t-SNE/UMAP to visualize.
- PCA can be used as preprocessing for classifiers with minimal accuracy loss.
